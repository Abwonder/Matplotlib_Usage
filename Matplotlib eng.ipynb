## The use of Matplotlib - Basic

```python
## importing the library
import matplotlib.pyplot as plt
```

```python
## This is to work with arrays
import numpy as np
```

```python
## creating dataset for this operations, random float dataset
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
## first basic plot in matplot
plt.scatter(x,y)  ## break down: first variable is horizontal axis(x-axis)
## second is the vertical (y-axis)
```

```python
plt.figure(figsize=(6,2)) ## size of the figure, first is length horizontally
## size and second is height length
plt.scatter(x,y , c = '#223BDC',alpha=.9,marker='X') ## c== colour, alpha ==
# transparency and marker == representation of the connecting points
plt.xlabel("this is my x axis") ## axis label
plt.ylabel("this is my y axis") ## axis label
plt.title("this is my x vs y")  ## head title on the plot
plt.grid()  ## the creationg to the quadrants
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y , '--r') 
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my sin wave graph")
plt.grid()
```

```python
plt.figure(figsize=(8,2))## size
plt.plot(x,y , '--b')  ## colour and point format, (marker)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my sin wave graph")
plt.grid()
```

```python
## Prep data for plotting
x = ['a', 'b' , 'c','d','e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
### Categorical vs numerical data
plt.figure(figsize=(5,3))
plt.bar(x,y,color = '#7E10D5')  ## i can use the code to change the colour!!
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```

```python
### tranpose is done using ".barh()" method
plt.figure(figsize=(5,3))
plt.barh(x,y,color = '#7E10D5')
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data,color="#700EBE")  ## it takes just one variable, uses freq 
plt.xlabel("representing uniqe val")
plt.ylabel("representing feq")
plt.title("Hist plot example")
plt.grid()
plt.savefig("test.png") ## saving the figure to directory (pwd)
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure() ## needed for creating 3D plot
ax = fig.add_subplot(projection = '3d')  ## for creating 3D plot
ax.scatter(x,y,z,c='#9327E8',alpha = .9)
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.show()
```
